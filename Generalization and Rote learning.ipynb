```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#create a toy dataset
x = np.linspace(-1, 1, 100)
signal = 2 + x + 2*x*x
```

```python
noise = np.random.normal(0,0.1, 100)
```

```python
y = signal + noise
```

```python
plt.plot(signal, 'b')
```

```python
plt.plot(y, 'g')
```

```python
plt.plot(noise, 'r')
```

```python
plt.xlabel("x")
```

```python
plt.ylabel("y")
```

```python
plt.legend(["without noise", "with noise", "noise"], loc =2)
plt.show()
```

```python
x_train = x[0:80]
y_train = y[0:80]
print("shape of x_train:", x_train.shape)
print("shape of y_train:", y_train.shape)
plt.show()
```
